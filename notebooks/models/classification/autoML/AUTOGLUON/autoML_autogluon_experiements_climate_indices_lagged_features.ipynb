```python
import autogluon as ag
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
dataset_train = TabularDataset(df_train)
```

```python
dataset_valid = TabularDataset(df_valid)
```

```python
dataset_test = TabularDataset(df_test)
```

```python
dataset_train
```

### path to save models and metrics 

```python
save_path = f'AG_observed_climate_indices_{period_features}_{max_features_lag}lags_{lag_to_realtime}lag_to_realtime_{region}_{varname_target}_{target_col}_{period_features}'  # specifies folder to store trained models
```

```python
predictor = TabularPredictor(label=target_col, path=save_path).fit(dataset_train, tuning_data=dataset_valid)
```

### loads the saved model 

```python
predictor = TabularPredictor.load(save_path) 
```

### evaluate over the independant test set (2017 - 2019)

```python
test_set_results = predictor.evaluate(dataset_test)
```

```python
test_set_results = pd.Series(test_set_results) 
```
